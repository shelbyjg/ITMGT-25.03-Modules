```python
import requests
import time
import matplotlib.pyplot as plt

def get_iss_location():
    # Get current ISS location
    url = "https://api.wheretheiss.at/v1/satellites/25544"
    response = requests.get(url)
    result = response.json()
    
    # Save the current ISS location in new variables
    latitude = float(result['latitude'])
    longitude = float(result['longitude'])
    timestamp = result['timestamp']
    return latitude, longitude, timestamp

# Collect data points
data_points = []
for i in range(100):
    latitude, longitude, timestamp = get_iss_location()
    data_points.append((latitude, longitude, timestamp))
    print(f"Data point {i+1}: Latitude={latitude}, Longitude={longitude}, Timestamp={timestamp}")
    time.sleep(10)  # Wait for 10 seconds before the next request

# Extract location for plotting
latitude, longitude, timestamp = zip(*data_points)

# Plotting the ISS path
plt.figure(figsize=(10, 6))
plt.plot(longitude, latitude, marker='o', linestyle='-')
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("International Space Station Path")
plt.grid(True)

# Display the plot
plt.show()

num_data_points = len(data_points)
print(f"Data collected: {num_data_points} points")
```
